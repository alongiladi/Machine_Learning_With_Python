## Your First Masterpiece: Plotting a Graph

Let's start by importing the `matplotlib` library. The standard way to do this is to import the `pyplot` module with the alias `plt`.

Now, let's create our first plot! We'll imagine we're tracking the stock price of a fictional company, 'DataCorp', over a week.

### Code Explanation

```python
stock_prices = [10, 12, 9, 15, 14, 18, 20]
plt.plot(stock_prices)
plt.show()
```
*   `stock_prices`: This is a simple list of numbers representing the stock price at the end of each day.
*   `plt.plot(stock_prices)`: This function creates a line plot of the data. By default, the x-axis will be the indices of the list (0, 1, 2, ...).
*   `plt.show()`: This function displays the plot.

### Code

**Real-world application:** In machine learning, you'll use line plots to visualize the performance of your model over time, such as the training loss or accuracy after each epoch.

````python
import matplotlib.pyplot as plt

stock_prices = [10, 12, 9, 15, 14, 18, 20]
plt.plot(stock_prices)
plt.show()
````

And just like that, you've created your first visualization! It's that simple to get started with Matplotlib.

**A quick note on Jupyter notebooks:**
> In a Jupyter notebook, you don't need to do any special setup. Importing `matplotlib.pyplot` is enough to make your plots appear directly in the notebook. In the past, you might have seen the command `%matplotlib inline`, but this is no longer necessary in modern Jupyter environments.

### Customizing Your Plot: Axes and Labels

The default plot is a great start, but we can make it much more informative. Let's add some labels and adjust the axes to give our data some context.

```python
days = [1, 2, 3, 4, 5, 6, 7]
stock_prices = [10, 12, 9, 15, 14, 18, 20]
plt.plot(days, stock_prices)
plt.axis([0, 8, 5, 25])
plt.show()
```
*   `plt.plot(days, stock_prices)`: This time, we're providing both the x-values (`days`) and the y-values (`stock_prices`).
*   `plt.axis([0, 8, 5, 25])`: This function sets the range of the axes with a list of `[xmin, xmax, ymin, ymax]`.

### Code

**Why this is important:** Controlling the scale and bounds of your plots is crucial for making your data easy to understand. You can use it to zoom in on a specific area of interest or to provide a broader context.

Let's give our graph a bit more room to breathe by adjusting the axes.

````python
days = [1, 2, 3, 4, 5, 6, 7]
stock_prices = [10, 12, 9, 15, 14, 18, 20]
plt.plot(days, stock_prices)
plt.axis([0, 8, 5, 25])
plt.show()
````

### Plotting a Mathematical Function

Matplotlib is also great for visualizing mathematical functions. Let's plot a simple parabola.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
plt.plot(x, y)
plt.show()
```
*   `import numpy as np`: We'll use NumPy to create a smooth range of x-values.
*   `x = np.linspace(-2, 2, 500)`: This creates an array of 500 numbers evenly spaced between -2 and 2.
*   `y = x**2`: We calculate the square of each x-value.
*   `plt.plot(x, y)`: We plot our parabola.

### Code

**Real-world application:** In machine learning, you might use this to visualize the cost function of your model, helping you to understand how the error changes as you adjust the model's parameters.

We'll use NumPy's `linspace` function to create a smooth curve.

````python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
plt.plot(x, y)
plt.show()
````

### Adding Titles, Labels, and a Grid

A plot without labels is like a book without a title. Let's make our plot more descriptive.

```python
plt.plot(x, y)
plt.title("The Parabola: A Beautiful Curve")
plt.xlabel("x-values")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```
*   `plt.title()`: Adds a title to the plot.
*   `plt.xlabel()`: Labels the x-axis.
*   `plt.ylabel()`: Labels the y-axis.
*   `plt.grid(True)`: Adds a grid to the plot for better readability.

### Code

**Why this is important:** Clear labels and a grid make your plots professional and easy to understand. This is essential when you're presenting your findings to others.

Let's make our plot publication-ready!

## Subplots: Multiple Plots in One Figure

Sometimes you want to display multiple plots in a single figure. Matplotlib makes this easy with subplots.

### Code Explanation

```python
x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2nd subplot
plt.plot(x, x**2)
plt.subplot(2, 2, 3)  # 3rd subplot
plt.plot(x, x**3)
plt.subplot(2, 2, 4)  # 4th subplot
plt.plot(x, x**4)
plt.show()
```
*   `plt.subplot(rows, cols, index)`: This function creates a grid of subplots and sets the specified one as active.

### Code

**Real-world application:** This is incredibly useful for creating dashboards where you want to see multiple visualizations at once. For example, you could show the distribution of different features of your dataset, all in one figure.

You can create a grid of subplots and draw on each one.

````python
x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2nd subplot
plt.plot(x, x**2)
plt.subplot(2, 2, 3)  # 3rd subplot
plt.plot(x, x**3)
plt.subplot(2, 2, 4)  # 4th subplot
plt.plot(x, x**4)
plt.show()
````

## Histograms: Understanding Your Data's Distribution

Histograms are a fundamental tool for understanding the distribution of your data.

### Code Explanation

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.hist(data, bins=10, rwidth=0.8)
plt.show()
```
*   `plt.hist()`: This function creates a histogram. `bins` controls the number of bins, and `rwidth` controls the width of the bars.

The data values go from **1** to **3.2**.

- With `bins=10`, Matplotlib divides this range `(3.2 - 1 = 2.2)` into **10 equal-width intervals**.  
- So each bin has a width of `2.2 / 10 = 0.22`.  
- Then:  
  - The first bin covers `[1.0, 1.22)`  
  - The second bin `[1.22, 1.44)`  
  - â€¦ and so on until `[3.0, 3.22]`  

### Code

**Real-world application:** Before you start building a machine learning model, you should always explore your data. Histograms are a great way to see if your data is skewed, if there are any outliers, or if it follows a particular distribution.

````python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.hist(data, bins=10, rwidth=0.8)
plt.show()
````
